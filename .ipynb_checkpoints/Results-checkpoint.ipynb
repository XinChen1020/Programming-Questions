# Task 1

```python
df_director = pd.read_csv('disney-director.csv')

df_director
```

```python
# Counting titles per director
titles_per_director = df_director.groupby('director')['name'].count().reset_index(name='number_of_titles')

# Ordering the table from highest to lowest number of titles directed
titles_per_director_sorted = titles_per_director.sort_values('number_of_titles', ascending=False)

titles_per_director_sorted
```
